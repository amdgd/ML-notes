```python
def gen_players(dataframe):
		'''docstring'''
		players = set(tuple(i for i in dataframe.P1.unique())
           +tuple(i for i in dataframe.P2.unique()))
		ids = [n+1 for n, pl in enumerate(sorted(players))]
		p1s = tuple(i for i in dataframe.P1)
		p2s = tuple(i for i in dataframe.P2)
		regions = tuple(i for i in dataframe.region)
		region_dict = {}
		for player in players:
		    for n in range(0,len(p1s)):
		        if  p1s[n] == player or p2s[n] == player:
		            region_dict[player] = regions[n]
		            break
		players = pd.DataFrame(sorted(players), columns=["name"])
		players["pid"] = pd.Series(sorted(ids))
		players["region"] = [region_dict[player] for player in players.name]
		return players

def gen_ids(dataframe):
	p1id = [player_dict[pl][0] for pl in dataframe.P1]
	p2id = [player_dict[pl][0] for pl in dataframe.P2]
	p1ids = tuple("{}{}".format(p1id[n],character_dict[ch]) for n, ch in enumerate(dataframe.Char1))
	p2ids = tuple("{}{}".format(p2id[n],character_dict[ch]) for n, ch in enumerate(dataframe.Char2))
	dataframe['winner_id'] = [p1ids[n] if p1s[n]==winner else p2ids[n] for n, winner in enumerate(dataframe.Winner)]
	dataframe['loser_id'] = [p2ids[n] if p1s[n]==winner else p1ids[n] for n, winner in enumerate(dataframe.Winner)]
	p1ids += (p2ids)
	player_ids = p1ids; del p1id, p2id, p1ids, p2ids
	unique_ids = set(player_ids)
	return player_ids, unique_ids

def update_match_ids(dataframe):
	dataframe['match_id'] = (["{}, {}".format(player_dict[p1s[n]][0], 
												player_dict[p2s[n]][0]) 
								for n in range(0, len(dataframe))])

def build_overall_table(dataframe):
	overall_record = pd.DataFrame(sorted(unique_ids), columns=["pid"])
	overall_record["player"] = [rev_player_dict[int(str(i)[:-2])][0] for i in overall_record.pid]
	overall_record["character"] = [rev_character_dict[int(str(i)[-2:])] for i in overall_record.pid]
	overall_record["region"] = [rev_player_dict[int(str(i)[:-2])][1] for i in overall_record.pid]
	wins = dataframe.winner_id.value_counts()
	losses = dataframe.loser_id.value_counts()

	overall_record['wins'] = get_count(wins, overall_record.pid)
	overall_record['losses'] = get_count(losses, overall_record.pid)
	overall_record['w/l'] = overall_record.wins - overall_record.losses

	overall_record['rank'] = find_closest_value(iterable=overall_record['w/l'], levels=[i for i in ranks["level"]], iterable_2=ranks['+/-'])
	overall_record.sort_values(by=['w/l', 'wins'], ascending=False).head()
	return overall_record

def append_winner_loser_rank(dataframe):
	ids = tuple(i for i in overall_record.pid)
	rankings = tuple(i for i in overall_record['rank'])
	rank_dict = {key : value for (key,value) in zip(ids, rankings)}
	dataframe['winner_rank'] = [rank_dict[id_no] for id_no in dataframe.winner_id]
	dataframe['loser_rank'] = [rank_dict[id_no] for id_no in dataframe.loser_id]
	dataframe['win_value'] = dataframe.loser_rank/dataframe.winner_rank + .5
	dataframe['loss_value'] = dataframe.winner_rank/dataframe.loser_rank + .5

def build_strength_table(dataframe):
	strength_record = pd.DataFrame(sorted(unique_ids), columns=["pid"])
	strength_record["player"] = [rev_player_dict[int(str(i)[:-2])][0] for i in strength_record.pid]
	strength_record["character"] = [rev_character_dict[int(str(i)[-2:])] for i in strength_record.pid]
	strength_record["region"] = [rev_player_dict[int(str(i)[:-2])][1] for i in strength_record.pid]
	def gen_strength_dict(ids, dataframe, id_col, val_col):
	    strength_dict = []
	    for id_no in ids:
	        frame = dataframe[dataframe[id_col] == id_no]
	        try:
	            strength_dict.append((id_no, int(frame[val_col].sum())))
	        except ZeroDivisionError:
	            strength_dict.append((id_no, 0))
	    return [i[1] for i in strength_dict]
	strength_record['winStrength'] = gen_strength_dict(strength_record.pid, dataframe, "winner_id", "win_value")
	strength_record['lossStrength'] = gen_strength_dict(strength_record.pid, dataframe, "loser_id", "loss_value")

	strength_record['w/l'] = strength_record.winStrength - strength_record.lossStrength
	strength_record['rank'] = find_closest_value(iterable=strength_record['w/l'], levels=[i for i in ranks["level"]], iterable_2=ranks['+/-'])
	strength_record.sort_values(by=['w/l', 'winStrength', 'lossStrength'], ascending=[False, False, True]).head()
	return strength_record

def gen_win_strength_chart():
	strength = strength_record.pivot("player", "character", "winStrength")
	fig = plt.figure(figsize=(12, 10))
	sns.heatmap(strength, 
	            annot=True, 
	            linewidths=.25,
	            cmap="Greens")
	plt.xticks(rotation=45, fontweight="bold")
	plt.xlabel("Character", fontsize=16)
	plt.ylabel("Player", fontsize=16)
	plt.title("Player Win Strength Matrix", 
	          fontsize=32, 
	          fontweight="bold")
	plt.savefig("charts/Player Win Strength Matrix.png")
	# plt.show()
	return 0

def gen_probability_charts(dataframe):
	winners = dataframe[["Winner","Loser","win_dummy"]].pivot_table(index="Winner", columns="Loser", values="win_dummy", aggfunc=sum)
	losers = dataframe[["Winner","Loser","win_dummy"]].pivot_table(index="Loser", columns="Winner", values="win_dummy", aggfunc=sum)
	probability = winners/(winners + losers)
	fig = plt.figure(figsize=(12, 10))
	sns.heatmap(probability, 
	            annot=True, 
	            linewidths=.5,
	            cmap="Greens")
	update_and_show_plot("Player Win Probability Matrix", save_fig="y")




	winners = (dataframe[["winChar","loseChar","win_dummy"]]
	           .pivot_table(index="winChar", 
	                        columns="loseChar", 
	                        values="win_dummy", 
	                        aggfunc=sum)
	           )
	losers = (dataframe[["winChar","loseChar","win_dummy"]]
	          .pivot_table(index="loseChar", 
	                       columns="winChar", 
	                       values="win_dummy", 
	                       aggfunc=sum))
	probability = winners/(winners + losers)*100


	fig = plt.figure(figsize=(12, 10))
	sns.heatmap(probability, 
	            annot=True, 
	            linewidths=.25, 
	            cmap="Greens")
	update_and_show_plot("Character Win Probability Matrix",
	                     save_fig="y")
	return 0

def export_data(dframe1, dframe2, dframe3):
	dframe1.to_csv("updates/Match Log.csv", index=False, header=True)
	(dframe2.sort_values(["w/l", "wins", "losses", "rank"], 
	                            ascending=[False,False,True, False])
	               .to_csv("updates/Overall Standings.csv", 
	                       index=False, 
	                       header=True))
	(dframe3
	 .sort_values(["w/l", "winStrength", "lossStrength"], 
	              ascending=[False,False, True])
	 .to_csv("updates/Strength Standings.csv", 
	         index=False, 
	         header=True))
	return 0
```
