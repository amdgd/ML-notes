```python
fake = Faker()
#Number of records to be generated
j = 0

#An empty list
actions = []

#Bgeinning loop
while (j <= 5005):
   
    r_score = int(fake.numerify('#'))
    f_score = int(fake.numerify('#'))
    m_score = int(fake.numerify('#'))
    avg_revenue_per_month =  revenue / random.randint(1,12)
    revenue = random.randint(500, 5870)
    discount_percentage = random.randint(10, 22)

                
    action = {
        "_index": INDEX,
        '_op_type': 'create',
        "_type": DOC_TYPE,
        "_id": fake.uuid4(),
        "_source": {
            "name": fake.name(),
            "email_unsubscribe": fake.boolean(),
            "fav_color":fake.color_name(),
            "revenue" : randint(100,9567),
            "country": fake.country(),
            "address": fake.address(),
            "email" : fake.safe_email(),
            "signup_date" : fake.date_time_this_year(),
            "job": fake.job(),
            "r_score": r_score,
            "f_score": f_score,
            "m_score": m_score,
            "rfm_score": r_score + f_score + m_score,
            "revenue": revenue,
            "avg_revenue_per_month": avg_revenue_per_month,
            "num_of_orders": int(revenue / avg_revenue_per_month),
            "discount_percentage": discount_percentage,
            "total_discount_revenue": revenue * discount_percentage /100,
            "lifecycle":random.randint(1, 10)
            }
        }
    
    actions.append(action)
    j += 1
    

    while (len(actions) > 50):
        helpers.bulk(es, actions)
        del actions[0:len(actions)]
        print("50 more records inserted...")
        # Empty the actions list after 10,000 times
        break

# Some data is remaining since the loop has gone when len(actions) is less than 10,000
if (len(actions) > 0):
    helpers.bulk(es, actions)
    del actions[0:len(actions)]
print("All records inserted")
   
```
