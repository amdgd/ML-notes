## Groupby in Panda

```python
#import pandas
import pandas as pd
```

```python
car_dataset_url ='https://raw.githubusercontent.com/ankitind/sample_datasets/master/car_ad.csv'
car_ads = pd.read_csv(car_dataset_url, header=0)
car_ads.head()
```

```python
car_ads.groupby('car')
```

```python
du_car_ads = car_ads
print(pd.get_dummies(du_car_ads['body']).head(5))
```

```python
car_ads.groupby('body').count()
```

```python
#df['preTestScore'].groupby(df['company'])
s = car_ads['engType'].groupby(car_ads['year'])
s.describe()
```

```python
#car_ads.groupby(['body', 'engType']).median()
car_ads.groupby(['body', 'engType']).count()

```

```python
# df.groupby(['regiment', 'company']).mean()
car_ads.groupby(['car', 'model']).mean().add_prefix('mean_')
```
