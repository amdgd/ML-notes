```python
df = car_ads
print(car_ads.dtypes)
char_cols = df.dtypes.pipe(lambda x: x[x == 'object']).index
print(char_cols)
label_mapping = {}

for c in char_cols:
    df[c], label_mapping[c] = pd.factorize(df[c])
    
df.info()
sns.lmplot(x='car', y='body', data = df, fit_reg=False)
plt.show()
```

```python

# Initialize the plot
fig = plt.figure(figsize=(14,2))
ax1 = fig.add_subplot(141)
ax2 = fig.add_subplot(142)
ax3 = fig.add_subplot(143)
ax4 = fig.add_subplot(144)
ax5 = fig.add_subplot(245)
ax6 = fig.add_subplot(246)
ax7 = fig.add_subplot(247)
ax8 = fig.add_subplot(248)
#ax4 = fig.add_subplot(122)
#ax4.hist(car_ads.price, bins=50)
#plt.hist(car_ads.engV, bins=50)
#sns.distplot(car_ads.year, ax=ax1)
ax1.set_title('Ax1 Histogar for Year')
ax1.hist(car_ads.year, bins=25)
#ax1.ylabel(rotation=0)

ax5.set_title('Ax1 Histogar for Year')
sns.distplot(car_ads.year,  bins=25, ax=ax5)

ax2.set_title('Ax2 Histogar for Price')
ax2.hist(car_ads.price, bins=10)


ax3.set_title('Ax3 Histogar for Mileage')
ax3.hist(car_ads.mileage, bins=10)

ax4.set_title('Ax4 engV for Mileage')
ax4.hist(car_ads.engV.dropna(), bins=10)

#plt.hist(car_ads.mileage, bins=50)
plt.show()
```
