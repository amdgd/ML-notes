### 2. One and Ones
We use one (singular) and ones (plural) to avoid unnecessary repetition.

- See those two girls? Helen is the tall one and Jane is the short one.
- Which is your car, the red one or the blue one?
- My trousers are torn. I need some new ones.
- See those two girls. Helen is the one on the left.
- Let’s look at the photographs. The ones you took in Paris.

**We often use them after Which ... in questions:**

- You can borrow a book. Which one do you want?
- There are lots of books here. Which ones are yours?

### 4.  Reciprocal pronouns 
when two or more people do the same thing.

### 5. Indefinite pronouns
We use indefinite pronouns to refer to people or things without saying exactly who or what they are. 

| somebody | someone | something |
| --- | --- | --- |
| anybody | anyone | anything |
| nobody | no one | nothing |
| everybody | everyone | everything|


We use a singular verb after an indefinite pronoun:

> 
- Everybody *loves* Sally.

When we refer back to an indefinite pronoun we normally use a plural pronoun:

> 
- Everybody enjoyed the concert. *They* stood up and clapped.
- I will tell somebody that dinner is ready. *They* have been waiting a long time.


## 6. Demonstrative Pronouns

That, this, these and those are demonstrative pronouns. They take the place of a noun or noun phrase that has already been mentioned.

> This is used for singular items that are nearby. These is used for multiple items that are nearby. The distance can be physical or metaphorical.

# Verb
Show action or state of being

## 1. Action verbs
    - Transistive verbs
    - Instransistive verbs
    
#### Transitive verb
A transitive verb has two characteristics. First, it is an action verb, expressing a doable activity like kick, want, paint, write, eat, clean, etc. Second, it must have a direct object, something or someone who receives the action of the verb.
- Sylvia kicked Juan under the table.
- Joshua wants a cookie
- My dog hears the car sound


#### InTransitive verb
An intransitive verb has two characteristics. First, it is an action verb, expressing a doable activity like arrive, go, lie, sneeze, sit, die, etc. Second, unlike a transitive verb, it will not have a direct object receiving the action.
- He sailed around the globe
- We arrived at the theater
- My dog hears well

## 2. Linking verbs (state-of-being)
The most common linking verbs are formed from the verbs 
- to be: 
> am, are, is, was, and were. 

Other words often used as linking verbs are 
> appear, become, grow, remain, seem, and the “sense” verbs: feel, look, smell, sound, and taste.


- I am happy
    - I = happy
- Rob is tired
    - Rob = tired

## 3. Auxillary/Helping verbs
To show additional difference in meaning, verbs often use helping words that suggest the time at which the action of the verb takes place and other kinds of meaning. These words are called helping/ auxiliary verbs, and they always come before the main verbs. Verbs that consist of helping verbs and a main verb are called verb phrases. Look carefully at the following sentences:

- I am going to work
    - am (helping)
    - going (action)
- I *will* go. 
- He *had* studied. 
- Zoie *did* not want lunch. 
- The sisters *were* saddened. 
- The child *was* photographed. 
- They might *have been* selected.

Common Helping verbs
> can, could may, might, must, ought shall, should, will, would have, has, had do, does, did am, is, are, was, were, been, be, being


Some verbs can be either helping/ auxiliary verbs or main verbs. In other words, if they appear alone without a helping verb/ auxiliary verb, they are main verbs. But, if they come before a main verb, they are helping/ auxiliary verbs or main verbs.

### Active vs Passive Voice
Active voice describes a sentence where the subject performs the action stated by the verb.
 - Tells what a person or a thing does
 - The subject performs the action and typically comes before the verb
 - The object is who or what that recieves the action and typically comes after the action (verb).
 

In passive voice sentences, the subject is acted upon by the verb. 
- Tell what is being done to someone or something. 
- The subject performs the action and typically comes after the action (verb). 
- The object is who or what receives the action and typically comes before the action (verb).



|           Active Voice           |           Passive Voice           |
|:-------------------------------- |:--------------------------------- |
| The cleaning crew (subj.) vaccums (verb) the office (obj.) every night.| The office (obj.) is vaccumed (verb) by the crew (subj.) every night.)|
    | Alex (subj.) posted (verb) the video (obj.) on Facebook. | The video (subj.) was posted (verb) on Facebook by Alex (subj.). |
| She (sub) faxed her (verb) application (obj.) for a new job | The applictaion (obj) for a new job was faxe (verb) by her (subj)|



## 4. Tense
Present, Past, and Future

There are four aspects of each tense. The term ‘aspect’ is connected with the term ‘tense’, but is independent of it. It gives a perspective on the verb, indicating whether the action is complete, or is still going on.

- Simple : for recurring action
- Continuous : for ongoing action at precise moment
- Perfect : for completed past action leadin to current events
- Perfect Continuous : for action that has been or had been or will have been going on, until now or then but ends, or has ended or will end

## 5. Participle
A participle is a word which is partly a verb and partly an adjective.
The three types of participles are 
- the present participle (ending ing) and 
- the past participle (usually ending -ed, -d, -t, -en, or -n).
- the perfect participle (having + past participle)

| The Verb | The Present Participle | The Past Participle | The Perfect participle |
| -------- | ---------------------- | ------------------- | ---------------------- |
| to rise  | the rising runs |  the risen sun | having risen |
| to boil | the boiling water | the boiled water | having boiled |
| to break | the breaking news | the broken news | having broken |




### Participle Phrases
It is really common to see participles in participle phrases. A participle phrase also acts like an adjective. In the examples below, the participle phrases are shaded and the participles are in bold:

- The man **carrying** the bricks is my father.
    - (The participle phrase carrying the bricks describes the the man.)
    
    
- She showed us a plate of scones **crammed** with cream.
    - (The participle phrase crammed with cream describes the scones.)
    

- **Whistling** the same tune as always, Ted touched the front of his cap with his forefinger as she dismounted.
    - (The participle phrase Whistling the same tune as always describes Ted.)
    

- **Stunned** by the blow, Mike quickly gathered his senses and searched frantically for the pepper spray.
    - (The participle phrase Stunned by the blow describes Mike.)
