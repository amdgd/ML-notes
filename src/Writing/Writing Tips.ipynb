# 1. Very 
Avoid using very


# 2. Too many adverbs and adjectives
When adjectives and adverbs are used too liberally, it slows down the pace of the narrative.

- Instead of: He **ran hurriedly** across the **dangerous** highway
- Use: He sprinted across the highway


- Instead of: She ate the sandwich quickly.
- Use: She devoured the sandwich.


- Instead of: The dark, dreary house had an empty, suspicious feel to it, the thick air stale and sour with undefined, scary kitchen odors.
- Use: The house had an empty feeling to it, the air stale with undefined kitchen odors.

# 3. Avoid adverbs that end in -ly 

- Instead of: “Why don’t you come over here and sit by me?” she asked **flirtatiously**.
- Use: “Why don’t you come over here and sit by me?” she asked battlingher eyelashes. 


- Instead of: The stone sank quickly.
- Use: The stone sank. (There is no need to emphasis on quickly)


- Instead of: She zoomed around the oval speedily
- Use: She zoomed arund the oval. 


- Instead of: He whispered to her **lovingly** … my sweet, dear lover, my angel … he purred his contentment, his joy …
- Use: He whispered words of love … my sweet, dear lover, my angel … he purred his contentment, his joy …

# 4. Using adverbs properly
Sometime using an adverb is necessary to convey the meaning. 

- Congress **recently** passed a new law.
- She entered the room **silently**.
- He drives a **dark** green sedan. 

# 5. Write strong prose
Make your proses tight and emphatic. Your phrases should never dangle or be superfluous.
Avoid prettyfying the proses. Be direct.

- Instead of: Find out when a weak verb *is* the culprit
- Use: Find out when a weak verb *creates* a problem


- Weak: He is a student who is intelligent and confident. He is always completing assignments on time.
- Stronger: The intelligent, confident student always completes assignments on time.


- Weak: Bill is going to start bringing all of his books to the group session for studying.
- Stronger: Bill decided to bring all of his books to the group study session.


- Weak: One key to **being** successful college student **is** that they **are** good at managing their time and **are** productive workers. 
- Stronger: Successful college students **manage** their time effectively and **work** productively.



- Weak: This key **is** a benefit because it **is** helpful in guarding students from **being** distracted like so many students **are**.
- Stronger: This key to success **guards** the students from distraction that **drag** their classmates down.


- Weak: College students who **are** successful **are** more likely to be engaged in their classes and to **be** more interested in course content.
- Stronger: Successful college students engage more in their classes and show interest in course content.

### Strategies to reduce 'be' verbs or a "be verb phrase"
#### 1. Replace them with a action verb 

- Weak: The book **was** interesting. 
- Stronger: The book **fascinated** me. 

- Weak: The sandwiches **were** eaten by the children. 
- Stronger: The children **devoured** the sandwich. 


#### 2. Move an adjective before a noun to avoid "be" descriptions.
- Weak: These puppies **are** adorable.
- Stronger: These adorable puppies...



#### 3. Tranform a noun or an adjective into an action verb
- Weak: He was the writer of the book.
- Stronger: He wrote the book.



## 7. Avoid verbs that replies on adverbs
Powerful verbs are strong enough to stand alone.

Examples

- The fox ~~ran quickly~~ dashed through the forest.
- She ~~menacingly looked~~ glared at her rival.
- He ~~secretly listened~~ eavesdropped while they discussed their plans.


## 8. Avoid verbs with -ing suffixes

Examples

- Before: He was walking…
- After: He walked…

 

- Before: She was loving the idea of…
- After: She loved the idea of…

 

Before: The family was starting to gather…

After: The family started to gather…
